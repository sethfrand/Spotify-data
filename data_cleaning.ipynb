```python
import pandas as pd
import seaborn as sns

spotify2 = pd.read_csv("spotify2.csv")
spotify2.head()
```

```python
for col in spotify2.columns:
    print(spotify2[col].isnull().sum())
```

```python
spotify2.drop("preview_url", axis=1, inplace=True)
spotify2.drop("track_id",axis =1, inplace=True)
spotify2.drop("track_url",axis=1,inplace=True)
spotify2.drop("playlist_name", axis=1, inplace=True)
spotify2.drop("artist_id",axis=1,inplace=True)
spotify2.drop("added_at",axis=1,inplace=True)
spotify2.drop("track_name",axis=1,inplace=True)
```

```python
spotify2
```

```python
spotify2.describe()
```

```python
spotify_encoded = pd.get_dummies(spotify2, columns=["is_explicit"],drop_first=True)
spotify_encoded
```

```python
numerical_data = spotify_encoded.select_dtypes(include=["int64","float64"])
spotify_encoded["is_explicit_True"] = spotify_encoded["is_explicit_True"].replace({True:1,False:0})

spotify_encoded
```

```python
import matplotlib.pyplot as plt
plt.scatter(spotify_encoded["popularity"],spotify_encoded["duration_min"])
plt.xlabel("Popularity")
plt.ylabel("Duration in minutes")
plt.title("Popularity vs Duration")
plt.savefig("Popularity-Duration.png")
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_explicit_True', y='popularity', data=spotify_encoded)
plt.xlabel('Explicit Content (0=Clean, 1=Explicit)')
plt.ylabel('Popularity')
plt.title('Popularity Distribution by Explicit Content')
plt.savefig("Popularity-Explicit_content.png")
```

```python
plt.figure(figsize=(8, 6))
spotify_encoded.groupby('is_explicit_True')['popularity'].mean().plot(kind='bar')
plt.xlabel('Explicit Content (0=Clean, 1=Explicit)')
plt.ylabel('Average Popularity')
plt.title('Average Popularity by Explicit Content')
plt.xticks(rotation=0)
plt.savefig("Average-Popularity-Explicit_content.png")
```
